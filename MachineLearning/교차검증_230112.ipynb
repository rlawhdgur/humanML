```python
import pandas as pd

wine = pd.read_csv('https://bit.ly/wine_csv_data')
wine.head()
```

## 데이터 셋 분리

```python
# 독립변수
data = wine[['alcohol', 'sugar', 'pH']].to_numpy()

# 종속변수
target = wine['class'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
                                                                                 # 실험 재현성
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size = 0.2, random_state =  42)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size = 0.2, random_state =  42)

x_train.shape, x_val.shape, y_train.shape, y_val.shape
```

## 모형 만들기


```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state = 42)
dt.fit(x_train, y_train)
```

```python
dt.score(x_train, y_train)
```

```python
dt.score(x_val, y_val)
```

## 교차검증


```python
from sklearn.model_selection import cross_validate

scores = cross_validate(dt, x_train, y_train)
print(scores)
```

```python
import numpy as np

print(np.mean(scores['test_score']))
```

## 데이터 시각화
- 각 칼럼별 ~ 종속변수
 - 박스플롯, 산점도, 히스토그램 등등 ,, 다양한 그래프를 그립니다.

## 모형 개발 코드 작성
- 파이프라인 구축
- pycaret 모형 만들때는, 내부적으로는 파이프라인 통해서 구현


```python
from sklearn.pipeline import Pipeline

# NA, NULL.. 결측지 값에 특정 값을 추가하는 방법(가장 단순한)
from sklearn.impute import SimpleImputer

# 수치 데이터를 변환
# 최소와 최대가 0 ~ 1 사이
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder

# 
from sklearn.compose import ColumnTransformer

# 숫자로 된 칼럼
num_features = ['Age', 'Fare', 'SibSp', 'Parch']
# 문자로 된 칼럼
cat_features = ['Embarked', 'Sex', 'Pclass']
# 필요 없는 칼럼
rem_features = ['Name', 'Cabin', 'Ticket', 'PassengerId']

# 각 데이터 타입에 맞는 transformer 구성
# 수치 데이터 결측지 값 채우는 방법 : 평균 또는 중간값
# 중간값으로 넣기
num_transformer = Pipeline(steps = [
    ('imputer', SimpleImputer(strategy = 'median')),
    ('scaler', MinMaxScaler())
])

# 문자 데이터 결측지 값 채우는 방법 : 가장 빈도수가 많은 데이터를 추가
cat_transformer = Pipeline(steps = [
    ('imputer', SimpleImputer(strategy = 'most_frequent')),
    ('onehot', OneHotEncoder())
])

# # 데이터 전처리기 파이프라인 구성
# preprocessor = ColumnTransformer(
#     transformer = [
#         ('numeric_features', num_transformer, num_features),     # 수치 데이터 변환기
#         ('category_features', cat_transformer, cat_features),     # 문자 데이터 변환기
#         ('remove_features', 'drop', rem_features)               # 불필요한 컬럼 제거
#     ]
# )

# 데이터 전처리기 파이프라인 구성
preprocessor = ColumnTransformer(
    transformers = [
        ('numeric_features', num_transformer, num_features), # 수치 데이터 변환기  
        ('category_features', cat_transformer, cat_features), # 문자 데이터 변환기 
        ('remove_features', 'drop', rem_features)  # 불필요한 컬럼 제거
    ]
)

```

```python
# 독립변수, 종속변수
target = train_df['Survived'].copy()
train_df = train_df.drop(['Survived'], axis = 1)

train_df.shape, target.shape
```

```python
# 로지스틱 회구분석 (-- 딥러닝 교육할 때 다시 배울 예정)
from sklearn.linear_model import LogisticRegression

# 모델 파이프라인 연결
model = Pipeline([
    ('preprocessor', preprocessor),
    ('clf', LogisticRegression(solver = 'liblinear'))
])

model.fit(train_df, target)
```

```python
accuray_score = round(model.score(train_df, target) * 100, 2)
accuray_score
```

## 모형 예측 및 csv 파일 변환


```python
y_pred = model.predict(test_df)
# y_pred

submission = pd.DataFrame({
    'PassengerId' : test_df['PassengerId'],
    'Survived' : y_pred
})


submission.to_csv(DATA_PATH + '/submission.csv', index = False)
```
